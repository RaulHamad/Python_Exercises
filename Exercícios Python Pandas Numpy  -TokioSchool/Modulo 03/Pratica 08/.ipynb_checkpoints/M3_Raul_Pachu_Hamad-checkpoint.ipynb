No exercício anterior, ao fazer um print as notas e nomes, vamos obter a nota e nome do aluno. Como pode ser considerado um problema de segurança, vamos encapsular o código para que não se possa aceder diretamente às variáveis e o acesso tenha de ser através de métodos.

````python
# Realizaro exercício aqui
class Aluno:
    
    __nome = None
    __nota = 0
    #construtor da classe Aluno
    def __init__(self,nome,nota):
        self.__nome = nome
        self.__nota = nota
        print("Aluno cadastrado com sucesso!")
        
    @property
    def nome(self):
        return self.__nome
    @nome.setter
    def nome(self,nome):
        self.__nome = nome
        
    @property
    def nota(self):
        return self.__nota
    @nota.setter
    def nota(self,nota):
        self.__nota = nota
        
        
    #método que deevolve uma string
    def __str__(self):
        return f'O aluno {self.__nome} tirou nota {self.__nota}.'
    #método que verifica se o aluno está aprovado
    def qualificacao(self):
        if self.nota >= 7:
            return f'O aluno {self.__nome} está aprovado.'
        else:
            return f'O aluno {self.__nome} está reprovado.'
        

aluno_1 = Aluno("makunouchi", 5)
aluno_2 = Aluno("Takamura", 7.5)
#print(aluno_1.__nome)AttributeError
#print(aluno_2.__nota)AttributeError
aluno_1.__nome = "toto"
print(aluno_1.__nome)#altera o variavel da classe mas não altera construtor e metodos devido ao encapsulamento
print(aluno_1.__str__())#Só altera  acessando o setter da instancia
print(aluno_1)
print(aluno_1.nota)
aluno_1.nota = 10
print(aluno_1.nota)
print(aluno_1)
print(aluno_2)
print(aluno_2.nome)
aluno_2.nome = "Camus"
print(aluno_2.nome)
print(aluno_2)

````

# Exercício 6

**Exercício sobre pontos, coordenadas e vetores no plano cartesiano. A seguir os conceitos básicos.**

### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./eixos.png"  />



### Pontos e coordenadas

O objetivo é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Puntos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vector.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que significa:

5 à direita e 2 para cima.

#### Experiência

````python
# Fazer o exercício aqui
import math

class Ponto:

    def __init__(self,coordenada_x=0, coordenada_y=0):
        self.coordenada_x = coordenada_x
        self.coordenada_y = coordenada_y
        print(f'Coordenada {self.coordenada_x,self.coordenada_y} criada')

    def __str__(self):
        return f'({self.coordenada_x},{self.coordenada_y})'

    def quadrante(self):
        """
        Método para verificar em que quadradante o ponto se encontra
        em um plano cartesiano

        """
        if self.coordenada_x == 0 and self.coordenada_y == 0:
            return f'O ponto {self.__str__()} está na origem'
        elif self.coordenada_x > 0 and self.coordenada_y > 0:
            return f'O ponto {self.__str__()} está no 1º quadrante'
        elif self.coordenada_x < 0 and self.coordenada_y > 0:
            return f'O ponto {self.__str__()} está no 2º quadrante'
        elif self.coordenada_x > 0 and self.coordenada_y < 0:
            return f'O ponto {self.__str__()} está no 3º quadrante'
        elif self.coordenada_x < 0 and self.coordenada_y < 0:
            return f'O ponto {self.__str__()} está no 4º quadrante'

    @classmethod
    def vetor(cls,ponto_x,ponto_y):
        """
        Calcular vetor resultante entre dois pontos
        ponto_x: A(x,y)
        ponto_y: B(x,y)
        return: (Bx-Ax),(By-Ay)
        """
        calculo_coord_x = ponto_y.coordenada_x - ponto_x.coordenada_x
        calculo_coord_y = ponto_y.coordenada_y - ponto_x.coordenada_y

        return f'({calculo_coord_x},{calculo_coord_y})'

    @classmethod
    def distancia(cls, ponto_x,ponto_y):
        """
        Calcula distancia entre dois pontos em um plano cartesiano
         ponto_x: recebe um ponto
         ponto_y: recebe outro ponto
        :return: valor da distancia entre os dois pontos
        """
        return f'{math.sqrt(((ponto_y.coordenada_x - ponto_x.coordenada_x)**2) + ((ponto_y.coordenada_y - ponto_x.coordenada_y)**2))}'


class Retangulo:

    def __init__(self,inicial=0,final=0):
        self.inicial = inicial
        self.final = final

    @classmethod
    def base(cls,ponto_x,ponto_y ):
        """
        Método que calcula o comprimento da base de um retangulo após receber dois pontos
        ponto_x: recebe um ponto x(X,Y)
        ponto_y: recebe um ponto y(X,Y)
        :return: comprimento da base
        """
        base_retangulo = ponto_y.coordenada_x - ponto_x.coordenada_x
        return abs(base_retangulo)

    @classmethod
    def altura(cls, ponto_x, ponto_y):
        """
        Método que calcula o comprimento da altura de um retangulo após receber dois pontos
        ponto_x: recebe um ponto x(X,Y)
        ponto_y: recebe um ponto y(X,Y)
        :return: comprimento da altura
        """
        altura_retangulo = ponto_y.coordenada_y - ponto_x.coordenada_y
        return abs(altura_retangulo)

    @classmethod
    def area(cls,ponto1, ponto2):
        """
        Método que calcula a área de um retangulo

        """

        return Retangulo.base(ponto1,ponto2)*Retangulo.altura(ponto1,ponto2)
print("-----Criando objetos A,B,C e D-----")
print(" ")
a = Ponto(2,3)
b = Ponto(5,5)
c = Ponto(-3,-1)
d = Ponto(0,0)
print(" ")
print("-----Verificando os quadrantes dos  objetos A,B,C e D-----")
print(" ")
print(a.quadrante())
print(b.quadrante())
print(c.quadrante())
print(d.quadrante())
print(" ")
print("-----Calculando vetor resultante-----")
print(" ")
print(f'O vetor resultante de A e B é: {Ponto.vetor(a,b)}')
print(f'O vetor resultante de B e A é: {Ponto.vetor(b,a)}')
print(" ")
print("-----Calculando a distântica entre os pontos-----")
print(" ")
print(f'A distancia entre os pontos A e B é de: {Ponto.distancia(a,b)}')
print(f'A distancia entre os pontos B e A é de: {Ponto.distancia(b,a)}')
print(" ")
print("-----Verificando qual dos pontos está mais Distante do ponto (0,0)-----")
print(" ")
dist_A = Ponto.distancia(a,d)
dist_B = Ponto.distancia(b,d)
dist_C = Ponto.distancia(c,d)



if max((dist_A),(dist_B),(dist_C)) == dist_A:
    print(f'O ponto mais distante da origem é o ponto A {a.__str__()} ')
elif max((dist_A),(dist_B),(dist_C)) == dist_B:
    print(f'O ponto mais distante da origem é o ponto B {b.__str__()} ')
else:
    print(f'O ponto mais distante da origem é o ponto C {c.__str__()} ')
print(" ")
print("-----Verificando comprimentos de um retangulo e calculando sua área-----")
print(" ")

#ULTIMA CORREÇÃO: inseri a função abs() para gerar o modulo e tirar os valores negativos de base, altura e area.

print(f'O comprimento da base entre os pontos A e B mede: {Retangulo.base(a,b)}')
print(f'O comprimento da altura entre os pontos A e B mede: {Retangulo.altura(a,b)}')
print(f'A área do retangulo formado pelos pontos A e B é de: {Retangulo.area(a,b)}')
````

# Exercício 9

#### Criação

Trabalhar com herança. Continuemos a expandir-nos em relação ao exercício anterior. Copiar todo o código que já possui e adicionar o seguinte:

* Um método chamado **filtro** ,que receba 2 parâmetros, a lista de veículos anteriomente construida e o nome de um dos tipos de veículos (Carro, Camião, etc..). O método deve percorrer a lista, procurar o nome do tipo da classe de cada objecto e encontrar os iguais ao argumento nome, e guardá-las numa nova lista. Esta nova lista deve ser devolvida ao programa principal onde será atravessada e o seu conteúdo exibido. A chamada no programa principal deve ter este aspeto: **``listaCarros = filtrar(lista, “Carro”)``**

#### Experiência

* Cria alguns veículos (mais do que um por tipo)
* Filtrar por diferentes tipos de veículos e mostrar os resultados devolvidos.

````python
# Fazer o exercício aqui
class Veiculo:
    
    def __init__(self,cor,rodas):
        self.cor = cor
        self.rodas = rodas
        
    def __str__(self):
        return f'Cor: {self.cor}\nRodas: {self.rodas}'
    @classmethod    
    def filtrar(cls,lista,nome_veiculo):
        lista_carros =[]
        for i in lista:
            if(type(i).__name__) == nome_veiculo:
                lista_carros.append(i)
        return lista_carros
    
    
class Carro(Veiculo):
    
    def __init__(self,cor,rodas,velocidade,cilindrada):
        super().__init__(cor,rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
    
    def __str__(self):
        return f'Cor: {self.cor}\nRodas: {self.rodas}\nCilindrada: {self.cilindrada}\nVelocidade: {self.velocidade}'
    

class Bicicleta(Veiculo):
    
    def __init__(self,cor,rodas,tipo):
        super().__init__(cor,rodas)
        self.tipo = tipo
        
    def __str__(self):
        return f'Cor: {self.cor}\nRodas: {self.rodas}\nTipo:  {self.tipo}'   
        
class Moto(Bicicleta):
    
    def __init__(self, cor,rodas,tipo,velocidade,cilindrada):
        super().__init__(cor,rodas,tipo)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
        
    def __str__(self):
        return f'Cor: {self.cor}\nRodas: {self.rodas}\nTipo:  {self.tipo}\nVelocidade: {self.velocidade}\nCilindrada: {self.cilindrada}'

class Camiao(Carro):
    
    def __init__(self,cor,rodas,velocidade,cilindrada,carga):
        super().__init__(cor,rodas,velocidade,cilindrada)
        self.carga = carga
        
    def __str__(self):
        return f'Cor: {self.cor}\nRodas: {self.rodas}\nCilindradas: {self.cilindrada}\nCarga: {self.carga}\nVelocidade: {self.velocidade}km/h'
    

carro1 = Carro("red",4,190,2000)
carro2 = Carro("green",4,160,1800)
carro3 = Carro("black",4,140,1400)
bicicleta1 = Bicicleta("green",2,"desportivo")
bicicleta2 = Bicicleta("pink",2,"urbano")
moto1 = Moto("blue",2,"urbano",150,600)
moto2 = Moto("gray",2,"desportivo",180,800)
moto3 = Moto("yellow",2,"urbano",220,1000)
camiao1 = Camiao("purple",4,90,3000,5)
camiao2 = Camiao("red",4,110,3500,8)

lista_veiculos = [carro1,carro2,carro3,bicicleta1,bicicleta2,moto1,moto2,moto3,camiao1,camiao2]

print("Filtrar Objetos do tipo Carro")
listaCarro = Veiculo.filtrar(lista_veiculos,"Carro")
contador = 1
for i in listaCarro:
    print(f'---Objeto {contador}---')
    print(i)
    contador +=1
    
print("Filtrar Objetos do tipo Moto e mostrar apenas sua cilindrada")
listaCarro = Veiculo.filtrar(lista_veiculos,"Moto")
contador = 1
for i in listaCarro:
    print(f'---Objeto {contador}---')
    print(i.cilindrada)
    contador +=1
contador = 1   
print("Filtrar Objetos do tipo Camiao e mostrar apenas sua carga")
listaCarro = Veiculo.filtrar(lista_veiculos,"Camiao")

for i in listaCarro:
    print(f'---Objeto {contador}---')
    print(i.carga)
    contador +=1
    
````
