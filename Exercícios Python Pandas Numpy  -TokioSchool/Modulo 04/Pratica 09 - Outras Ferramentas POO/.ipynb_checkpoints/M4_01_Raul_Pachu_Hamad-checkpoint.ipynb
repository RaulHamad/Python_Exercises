**1) Exercício de herança múltipla e polimorfismo**
* Criar a classe **Veiculos** com um construtor que inclua marca e modelo
* Esta classe **Veículos** também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer gasolina”
* Criar a classe **VEletricos** com um construtor que inclua marca, modelo e autonomia
* Esta classe VEletricos também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer eletricidade”
* Criar a classe **BicicletaEletrica** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **VEletricos** (já que é um veículo mais elétrico que normal)
* Criar a classe **Quadriciclo** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **Veículos** (já que apenas usa a eletricidade de modo pontual)
* Criar um objeto de **BicicletaEletrica** e outro de **Quadriciclo**
* Empregar as técnicas de polimorfismo aprendidas para conseguir que ao perguntar-lhe aos objetos que acabamos de criar por reabastecimento, ambos nos respondam adequadamente


```python
# Completar o exercício aqui
class Veiculos:
    def __init__(self,marca,modelo):
        self.marca = marca
        self.modelo = modelo
    
    def reabastecer(self):
        print("Este veículo tem de reabastecer gasolina")
    
class VEletricos:
    def __init__(self,marca,modelo,autonomia):
        self.marca = marca
        self.modelo = modelo
        self.autonomia = autonomia
        
    def reabastecer(self):
        print("Este veículo tem de reabastecer eletricidade")
        
class BicicletaEletrica(VEletricos,Veiculos):
    pass

class Quadriciclo(Veiculos,VEletricos):
    pass
        

bicicleta_eletrica_1 = BicicletaEletrica("scoot","desportivo",180)
bicicleta_eletrica_1.reabastecer()
print(bicicleta_eletrica_1.marca)

quadriciclo_1 = Quadriciclo("Toyota", "desportivo")
quadriciclo_1.reabastecer()
print(quadriciclo_1.marca)
print(bicicleta_eletrica_1.__dict__)
```

**2) Exercício de herança**
* Criar a superclasse **Pessoal_Universitario** com um único atributo que seja um dicionário que inclua (Keys): “id”, “nome” e “email” 
* Criar as classes **Escritorio**, **Professor** e **Aluno**. Todas herdaram de **Pessoal_Universitário** e os seus dados
* No caso de **Escritorio**, deve adicionar ao seu dicionário de dados o dado “Gabinete”
* No caso de **Professor**, deve adicionar ao seu dicionário de dados o dado “Especializacao”
* No caso de **Aluno**, deve adicionar ao seu dicionário de dados o dado “CreditosAprovados" (integer)  
* Criar métodos que sirvam para mostrar a informação de cada classe
* Criar objetos de todas as classes
* Mostrar a informação desses objetos e verificar o seu output.

```python
# Completar o exercício aqui


class Pessoal_Universitario:
    informacoes = {}
    
    def __init__(self,informacoes = {"id":None,"nome":None,"email":None}):
        self.informacoes = informacoes
        
    
    def __str__(self):
        
        return f'Classe Super\nID: {self.informacoes["id"]}\nNome: {self.informacoes["nome"]}\nEmail: {self.informacoes["email"]}'

class Escritorio(Pessoal_Universitario):
    def __init__(self,informacoes,gabinete):
        super().__init__(informacoes)
        self.informacoes = informacoes
        self.informacoes["gabinete"] = gabinete
        print("escritorio")
    
    def __str__(self):
        
        return f'ID: {self.informacoes["id"]}\nNome: {self.informacoes["nome"]}\nEmail: {self.informacoes["email"]}\nGabinete: {self.informacoes["gabinete"]}'
    
class Professor(Pessoal_Universitario):
    def __init__(self,informacoes,especializacao):
        super().__init__(informacoes)
        self.informacoes = informacoes
        self.informacoes["especializacao"] = especializacao
        print("Professor")
        
    def __str__(self):
        
        return f'ID: {self.informacoes["id"]}\nNome: {self.informacoes["nome"]}\nEmail: {self.informacoes["email"]}\nEspecialização: {self.informacoes["especializacao"]}'
    
class Aluno(Pessoal_Universitario):
    def __init__(self,informacoes, CreditosAprovados: int):
        super().__init__(informacoes)
        self.informacoes = informacoes
        self.informacoes["CreditosAprovados"] = CreditosAprovados
        print("Aluno")
    def __str__(self):
        
        return f'ID: {self.informacoes["id"]}\nNome: {self.informacoes["nome"]}\nEmail: {self.informacoes["email"]}\nCreditos: {self.informacoes["CreditosAprovados"]}'
    

pessoal_universitario_1 = Pessoal_Universitario({"id":111,"nome":"Lucas","email":"Lucas@python.com"})
print(pessoal_universitario_1)
print(" ")
pessoal_escritorio_1 = Escritorio({"id":444,"nome":"Matheus","email":"matheus@python.com"},45)
print(pessoal_escritorio_1)
print(" ")
pessoal_professor_1 = Professor({"id":555,"nome":"seiya","email":"seiya@python.com"},"mestrado")
print(pessoal_professor_1)
print(" ")
pessoal_aluno_1 = Aluno({"id":777,"nome":"Manoel","email":"Manoel@python.com"},10)
print(pessoal_aluno_1)
```

**3)  Escolher a temática que preferir e definir uma estrutura de classes e subclasses onde pratique a herança em pelo menos dois níveis (herança multinível). Criar uma estrutura de mínimo 5 classes, com os seus respetivos construtores e métodos. Criar alguns objetos e comprovar que a herança esteja a funcionar. Se puder, incluir a técnica de polimorfismo e herança múltipla.**

```python
# Completar o exercício aqui

class Telefone:
    def __init__(self,marca,cor):
        self.marca = marca
        self.cor = cor

    def iniciar_chamada(self):
        print("Discando...")
    
    def encerrar_chamada(self):
        print("Encerrando chamada...")
         
    def __str__(self):
        return f'Marca: {self.marca}\nCor: {self.cor}'

class CameraFotografica:

    def __init__(self,categoria):
        self.categoria = categoria

    def registrar_foto(self):
        print("tirando uma fotografia...")

    def __str__(self):
        return f'Categoria: {self.categoria}'

class Telemovel(Telefone):
    
    def __init__(self, marca, cor,operadora):
        super().__init__(marca, cor)
        self.operadora = operadora

    def enviar_sms(self):
        print("Enviando sms...")

    def __str__(self):
        return f'Marca: {self.marca}\nCor: {self.cor}\nOperadora: {self.operadora}'

class Tablet(Telemovel):
    def __init__(self, marca, cor, operadora,memoria):
        super().__init__(marca, cor, operadora)
        self.memoria = memoria
    def instalar_app(self):
        print("Instalando app...")

    def navegar_web(self):
        print("Navegando pela internet")

    def __str__(self):
        return f'Marca: {self.marca}\nCor: {self.cor}\nOperadora: {self.operadora}\nMemória: {self.memoria}Gb'


class Smartphone(Tablet,CameraFotografica):

    def __init__(self, marca, cor, operadora, memoria):
        super().__init__(marca, cor, operadora, memoria)
    
    def __str__(self):
        return f'Marca: {self.marca}\nCor: {self.cor}\nOperadora: {self.operadora}\nMemória: {self.memoria}Gb'

#objeto Telefone acede apenas os seus métodos
telefone_1 = Telefone("grambell","Preto")
print(f'Classe: {type(telefone_1).__name__}') 
print(telefone_1)
telefone_1.iniciar_chamada()
telefone_1.encerrar_chamada()
print(" ")

#objeto CameraFotografica acede apenas os seus métodos   
camera_fotografica_1 = CameraFotografica("Profissional")
print(f'Classe: {type(camera_fotografica_1).__name__}') 
print(camera_fotografica_1)
camera_fotografica_1.registrar_foto()
print(" ")

#objeto Telemovel acede os métodos de Telefone
telemovel_1 = Telemovel("Nokia","verde","Tim")
print(f'Classe: {type(telemovel_1).__name__}')
print(telemovel_1)
telemovel_1.iniciar_chamada()
telemovel_1.encerrar_chamada()
telemovel_1.enviar_sms()
print(" ") 

#objeto Tablet acede os métodos de Telefone e Telemovel
tablet_1 = Tablet("Samsung","vermelho","Tim",64)
print(f'Classe: {type(tablet_1).__name__}')
print(tablet_1)
tablet_1.iniciar_chamada()
tablet_1.encerrar_chamada()
tablet_1.enviar_sms()
tablet_1.instalar_app()
print(" ") 

#objeto Smartphone acede os métodos de Telefone, Telemovel, Tablet e CameraFotografica(prioridade a Tablet que está a esquerda)
smart_1 = Smartphone("xiaomi","cinza","meo",128)
print(f'Classe: {type(smart_1).__name__}')
print(smart_1)
smart_1.iniciar_chamada()
smart_1.encerrar_chamada()
smart_1.enviar_sms()
smart_1.instalar_app
smart_1.registrar_foto()
smart_1.navegar_web()
```
