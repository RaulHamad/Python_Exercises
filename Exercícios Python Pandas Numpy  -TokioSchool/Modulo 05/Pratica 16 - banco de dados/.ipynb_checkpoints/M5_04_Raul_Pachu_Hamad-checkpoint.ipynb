```python
import sqlite3

con = sqlite3.connect("biblioteca.db")
cur = con.cursor()
#Adicionando registros a tabela autor
sql_register = "INSERT INTO autor VALUES(?,?,?,?)"

register = [
    (1001,"Masami", "Kurumada", 1),
    (1002,"Mary", "Higgins Clark",0),
    (1003,"Akira","Toryama",1),
    (1004,"Alfonso","Sastre",0),
    (1005,"George","Morikawa",1)
]

for add in register:
    cur.execute(sql_register,add)
    
con.commit()

con.close()

```

```python
#Fazer as consultas das tabelas criadas

import sqlite3

con=sqlite3.connect("biblioteca.db")
cur = con.cursor()

sql_query_autor = "SELECT * FROM autor"
sql_query_livro = "SELECT * FROM livro"
sql_query_utilizador = "SELECT * FROM utilizador"

cur.execute(sql_query_autor)

dados = cur.fetchall()
print("----Listagem de autor-----")
for i in dados:
    print(i)

cur.execute(sql_query_livro)

dados = cur.fetchall()
    
print("----Listagem de livro-----")
for i in dados:
    print(i)
    
    
cur.execute(sql_query_utilizador)

dados = cur.fetchall()
    
print("----Listagem de utilizador-----")
for i in dados:
    print(i)
    
con.close()
```

```python
import sqlite3

con = sqlite3.connect("biblioteca.db")
cur = con.cursor()



# Listar todos os autores que estejam vivos (CLAUSULA WHERE)
list_autores_vivos = "SELECT nome,apelidos FROM autor WHERE estarVivo = 1"

cur.execute(list_autores_vivos)

dados_autor = cur.fetchall()
print("Listar todos os autores que estejam vivos")
for i in dados_autor:
    print(f'Nome: {i[0]}, Apelido: {i[1]}')
# Listar todos os livros que tenham sido escritos posteriormente a 1900

list_livro = "SELECT titulo, ano_escrito FROM livro WHERE ano_escrito > 1900"

cur.execute(list_livro)
dados_livro = cur.fetchall()
print(" ")
print("Listar todos os livros que tenham sido escritos posteriormente a 1900")
for i in dados_livro:
    print(f'titulo: {i[0]}, Escrito em: {i[1]}')
    
# Listar todos os utilizadores que se tenham levado mais de 10 livros e filtrar pelo nome

list_utilizador = "SELECT nome FROM utilizador WHERE numEmprestimos > 10"

cur.execute(list_utilizador)

dados_utilizador = cur.fetchall()
print(" ")
print("Listar todos os utilizadores que se tenham levado mais de 10 livros e filtrar pelo nome")
for i in dados_utilizador:
    print(f'Nome: {i[0]}')
```
