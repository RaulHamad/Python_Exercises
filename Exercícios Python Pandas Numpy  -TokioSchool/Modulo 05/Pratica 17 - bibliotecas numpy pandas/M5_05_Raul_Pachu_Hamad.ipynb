**1) Trabalhar com NumPy**

* Importar numpy e usar um alias para poder utilizar os seus recursos 
* Criar um array de elementos (de 10 a 100)
* Inverter o array
* Criar um array 3x3 com valores de 0 a 8 
* Criar um array identidade de 6x6
* Criar um array com valores aleatórios com o formato 3x3x3
* Criar dois arrays, com valores aleatórios, A e B, e verificar se são iguais (há várias opções, investigar)

```python
# Completar o exercício aqui
import numpy as np

#Criar um array de elementos (de 10 a 100)
array_100 = np.arange(10,101)
print("-----Criar um array de elementos de 10 a 100-----")
print(array_100)
print(" ")

#Inverter o array
array_invertido = np.flip(array_100)
print("-----Inverter o array-----")
print(array_invertido)
print(" ")

#Criar um array 3x3 com valores de 0 a 8
array_9_elementos = np.arange(9)

array_3x3 = np.reshape(array_9_elementos,[3,3])
print("-----Criar um array 3x3 com valores de 0 a 8-----")
print(array_3x3)
print(" ")

#Criar um array identidade de 6x6
array_identity = np.identity(6)
print("-----Criar um array identidade de 6x6------")
print(array_identity)
print(" ")

#Criar um array com valores aleatórios com o formato 3x3x3
array_aleatorio_3x3x3 = np.random.randint(27,size=(3,3,3))
print("-----Criar um array com valores aleatórios com o formato 3x3x3-----")
print(array_aleatorio_3x3x3)
print(" ")

# Criar dois arrays, com valores aleatórios, A e B, e verificar se são iguais (há várias opções, investigar)
array_1 = np.array([[2,4],[6,8]])
array_2 = np.array([[3,5],[7,9]])
print("-----Criar dois arrays, com valores aleatórios, A e B, e verificar se são iguais (há várias opções, investigar)-----")
print(array_1,array_2)
print(np.array_equal(array_1,array_2))
array_1 = np.array([[2,4],[6,8]])
array_2 = np.array([[2,4],[6,8]])
print(np.equal(array_1,array_2))
```

**2) Seguir com NumPy**

* Criar um array de 100 elementos aleatórios compreendidos entre 1 e 1000 (podem ser números inteiros). 
* Este array irá chamar-se dólares
* Fazer a conversão dos preços em dólares do primeiro array para um segundo array que se vai chamar euros. Devemos utilizar a  seguinte conversão: 
  - 1 USD = 0,874413 EUR
* Mostrar o resultado do array de euros, arredondando a 2 casas decimais.

```python
# Completar o exercício aqui
import numpy as np

"""dolares = np.linspace(1,1000,100).astype(int)

    quando uso o linspace iniciando pelo numero 1 , a funcao round() mostra um valor diferente quando colocar mais de uma casa decimal
"""
print("-----Criando array com 100 elementos aleatórios-----")
dolares = np.linspace(2,1000,100).astype(int)
print(dolares)
#conversão dolares para euro
converter = dolares*0.874413
euro = np.round(converter, decimals=2) #QUando coloco para mais de uma casa decimal, não me retorna o valor numero normal(ex.: 8.7000e-01)
print(euro.round(2))
```

* Mostrar quantos papéis fez “Robert de Niro” na sua carreira.
* Listagem de papéis como protagonista (n=1) que teve o ator “Charlton Heston” na década dos 60’s, ordenado  por ano de forma descendente.
* Mostrar quantos papéis para atores teve na década 50’s.
* Mostrar quantos papéis para atrizes teve na década 50’s.

```python
# Completar o exercício aqui
import pandas as pd

#Carregar como dataframe de pandas csv 05_05_imdb_títulos.csv e mostrar os seus 5 primeiros registos.
df_titulos = pd.read_csv("05_05_imdb_titulos.csv")
print("----- Os 5 primeiros registros de titulos-----")
print(df_titulos.head(5))
print(" ")


#Carregar como dataframe de pandas o csv 05_05_imdb_elenco.csv e mostrar os seus 5 primeiros registos
df_elenco = pd.read_csv("05_05_imdb_elenco.csv")
print("----- Os 5 primeiros registros de elenco-----")
print(df_elenco.head(5))
print(" ")


#Mostrar o número de registos do dataframe de títulos.
print(f'O numero de registros do arquivo csv titulos: {df_titulos.index.stop}')


#Mostrar o número de registos do dataframe de elenco.
print(f'O numero de registros do arquivo csv elenco: {df_elenco.index.stop}')


#Mostrar os 5 titulos mais antigos.
print(" ")
print("----- Os 5 Ultimos registros de titulos mais antigos-----")
titulos_antigo = df_titulos.sort_values("year")
print(titulos_antigo.head(5))


#Mostrar os filmes que no título contem a palavra “Drácula”. Quantos filmes são?
print(" ")
filme_dracula = df_titulos[df_titulos["title"].str.contains("Dracula")]
print("-----Título que contem a palavra “Drácula”-----")
print(filme_dracula)
print(f'Temos um total de {len(filme_dracula)} filmes com o nome Dracula')


#Mostrar os 10 títulos mais comuns (que mais se repetem).
print(" ")
print("-----Mostrar os 10 títulos mais comuns-----")
titulos_repetidos = df_titulos.groupby("title").size().sort_values(ascending=False)
print(titulos_repetidos.head(10))


#Mostrar qual foi o primeiro filme chamado “Romeo and Juliet”.
print(" ")
filme_romeoAndJuliet = df_titulos[df_titulos["title"] == "Romeo and Juliet"]
print("-----Primeiro filme chamado “Romeo and Juliet”-----")
print(filme_romeoAndJuliet["year"].sort_values(ascending=True).head(1))


#Listar todas os filmes que contenham a palavra “Exorcist”, ordenados do mais antigo ao mais recente.
print(" ")
filmes_exorcist = df_titulos[df_titulos["title"].str.contains("Exorcist")]
print("----- filmes que contenham a palavra “Exorcist”-----")
print(filmes_exorcist.sort_values("year"))


#Mostrar quantos filmes foram feitas no ano 1950.
print(" ")
filmes_1950 = df_titulos[df_titulos["year"] == 1950]
print("-----Mostrar quantos filmes foram feitas no ano 1950-----")
print(len(filmes_1950))


#Mostrar quantaos filmes foram feitas de 1950 a 1959 (inclusive).
print(" ")
filmes_1950_1959 = df_titulos[(df_titulos["year"] >= 1950) & (df_titulos["year"] <= 1959) ]
print("-----Mostrar quantaos filmes foram feitas de 1950 a 1959 (inclusive)-----")
print(filmes_1950_1959)
print(f'Foram feitos {len(filmes_1950_1959)} filmes entre 1950 e 1959')


#Mostrar todas os papéis que teve a trilogia “The Godfather”, agrupando pelo papel e forneca a
#informação da quantidade de vezes esse papel apareceu (exemplo: michael corleone aparece nos 3 filmes da triologia).
print(" ")
godfather = df_elenco[df_elenco["title"].str.contains("The Godfather")]
character = godfather.groupby("name").size().sort_values(ascending=False)
print("-----Papeis da pelicula 'Godfather'-----")
print(character)


#Mostrar o elenco completo, ordenado pela classificação “n” do filme “Drácula” de 1958
print(" ")
print("-----elenco completo, ordenado pela classificação “n” do filme “Drácula” de 1958-----")
elenco_dracula = df_elenco[(df_elenco["title"]==("Dracula")) & ((df_elenco["year"] == 1958))]
print(elenco_dracula.sort_values("n"))


#Mostrar quantos papéis teve “Bruce Wayne” como protagonista
#(Dica: Em Batman de 1989, existe Bruce Wayne e tambem Young Bruce Wayne, que deverão ambos aparecer).
print(" ")
print("-----Quantos papéis teve “Bruce Wayne” como protagonista-----")
papel_bruce_wayne = df_elenco[df_elenco["character"].str.contains("Bruce Wayne")]
print(papel_bruce_wayne.count())


#Mostrar quantos papéis fez “Robert de Niro” na sua carreira
print(" ")
print("-----Mostrar quantos papéis fez “Robert de Niro” na sua carreira-----")
papeis_robert = df_elenco[df_elenco["name"].str.contains("Robert De Niro")]
print(len(papeis_robert))


#Listagem de papéis como protagonista (n=1) que teve o ator “Charlton Heston” na
#década dos 60’s, ordenado por ano de forma descendente.
print(" ")
print("-----Listagem de papéis como protagonista que teve o ator “Charlton Heston” ordenado por ano de forma descendente.-----")
papeis_charlton = df_elenco[(df_elenco["name"].str.contains("Charlton Heston")) & (df_elenco["n"]==1) & (df_elenco["year"] >= 1960) & (df_elenco["year"] <= 1969) ]
print(papeis_charlton.sort_values("year"))


#Mostrar quantos papéis para atores teve na década 50’s.
print(" ")
print("-----Mostrar quantos papéis para atores teve na década 50’s.-----")
papeis_atores = df_elenco[(df_elenco["type"].str.contains("actor")) & (df_elenco["year"] >= 1950) & (df_elenco["year"] <=1959)]
print(len(papeis_atores))


#Mostrar quantos papéis para atrizes teve na década 50’s.
print(" ")
print("-----Mostrar quantos papéis para atores teve na década 50’s.-----")
papeis_atrizes = df_elenco[(df_elenco["type"].str.contains("actress")) & (df_elenco["year"] >= 1950) & (df_elenco["year"] <=1959)]
print(len(papeis_atrizes))
```
