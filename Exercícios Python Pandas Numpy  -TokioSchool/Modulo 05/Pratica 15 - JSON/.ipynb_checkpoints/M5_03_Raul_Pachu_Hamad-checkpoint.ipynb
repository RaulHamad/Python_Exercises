# JSON (JavaScript Object Notation)

1) Tenha em consideração o seguinte JSON String:

{"chefe_projeto":{"Nome":"Juan","Idade":28,"Experiência": ["Gestão","Finanças","Basesdedados"],"Residência": 
"Madrid","HorasProjeto":3500},"funcionários":[{"Nome":"Elena","Idade":26,"Experiência":
["JavaScript","Python"],"Residência":"Madrid","HorasProjeto":500},{"Nome":"Luis","Idade":31,"Experiência": 
["Django","Flask","Pyramid"],"Residência":"Barcelona","HorasProjeto":1100}]}


* Se necessário, utilizar um visualizador de JSON para o entender.
* O objetivo é criar um dicionário com todos os dados e estruturas internas necessárias para que seja igual ao JSON String.
* Criar estruturas mais pequenas até que se chegue a formar o mesmo JSON string que vemos acima.
* Armazenar esta estrutura (o dicionário) numa variável.
* Comprovar o seu tipo e mostrar por ecrã.
* Criar duas variáveis: horas_funcionários e horas_chefe
* Extrair as horas do chefe e introduzir na sua variável.
* Criar um ciclo que percorra a totalidade de funcionários que tem na estrutura e para cada funcionário somar as suas horas na sua variável correspondente.
* Somar todas as horas e imprimir.

```python
# Completa o exercicio aqui
import json

experiencia_juan = ["Gestão","Finanças","Basesdedados"]
experiencia_elena =  ["JavaScript","Python"]
experiencia_luiz = ["Django","Flask","Pyramid"]
dados_chefe = {"Nome":"Juan","Idade":28,"Experiência": experiencia_juan,"Residência": "Madrid","HorasProjeto":3500}
chefe_projeto = dados_chefe
dados_elena = {"Nome":"Elena","Idade":26,"Experiência": experiencia_elena,"Residência":"Madrid","HorasProjeto":500}
dados_luiz = {"Nome":"Luis","Idade":31,"Experiência": experiencia_luiz,"Residência":"Barcelona","HorasProjeto":1100}
funcionarios =[dados_elena,dados_luiz]
#Armazenando estrutura dicionario numa variavel:
estrutura_dict = {"chefe_projeto":dados_chefe,"funcionários":funcionarios}
#Comprovando tipo:
print(type(estrutura_dict))
print(estrutura_dict)
#Criando variaveis para extrair as horas:
    #Variavel chefe:
horas_chefe = estrutura_dict["chefe_projeto"]["HorasProjeto"]

print(" ")
    #variavel funcionarios:
horas_funcionarios = []
for horas in estrutura_dict["funcionários"]:
    horas_funcionarios.append(horas["HorasProjeto"])
print("-----Somatório das horas de projeto-----")
print(f'O chefe de projeto tem {horas_chefe} horas de projeto')
print(f'A soma de horas de projeto dos funcionários sao {sum(horas_funcionarios)}')
print(f'A soma de todas as horas totaliza {sum(horas_funcionarios,horas_chefe)}')
```

2) No exercício anterior trabalhou-se com um dicionário que nós mesmos/as criamos. Converter para um formato JSON String, imprimir o seu tipo e os seus dados. Para que serve este formato?

```python
# Completa o exercicio aqui
import json

json_string = json.dumps(estrutura_dict)
print("-----Tipo de dado-----")
print(type(json_string))
print(" ")
print(json_string)

#JSON é um formato para dados em formato de texto para que seja enviado de um sistema a outro.
#Muito utilizado por ser lido por varias linguagens de programação
```

```python
# Completa o exercicio aqui
import json

superherois = {
	"nomedaEquipa": "Super Hero Squad",
	"cidade": "Metro City",
	"formado": 2016,
	"baseSecreta": "Super Tower",
	"activo": "Sim",
	"membros": [
		{
			"nome": "SuperMan",
			"idade": 29,
			"identidadeSecreta": "Clart Kent",
			"poderes": [
				"Super força",
				"Super velocidade",
				"Raio laser"
			]
		},
		{
			"nome": "Batman",
			"idade": 350,
			"identidadeSecreta": "Bruce Wayne",
			"poderes": [
				"Detective",
				"Dinheiro",
				"Raio laser"
			]
		},
		{
			"nome": "Wonder Woman",
			"idade": 900,
			"identidadeSecreta": "Diana de Temiscira",
			"poderes": [
				"Super força",
				"Super velocidade"
			]
		}
	]
}

print(type(superherois))
print(superherois)
print(" ")
#-----Ao Superman falta-lhe como poder Voar-----
superherois["membros"][0]["poderes"].append("Voar")
print(superherois)
print(" ")
#-----No Batman, a idade é 35, não 350-----

superherois["membros"][1]["idade"]= 35
print(superherois)
print(" ")
#-----No Batman, retire-lhe o poder de "Raio laser"-----

superherois["membros"][1]["poderes"].remove("Raio laser")
print(superherois)
print(" ")
#----- No Wonder Woman falta-lhe o poder "Laço da verdade"-----
superherois["membros"][2]["poderes"].append("Laço da verdade")
print(superherois)

#Converter para JSON:

superherois_correcao = json.dumps(superherois)
print(" ")
print(type(superherois_correcao))
print(" ")
print(superherois_correcao)
```

4) Com base no exercício anterior, modificar a estrutura de **super_herois** para conseguir que **membros** tenham dois ramos: "membrosAtivos" e "membrosInativos", onde cada um destes ramos, armazenem os heróis que estão no ativo e os que não. Neste caso, introduzimos SuperMan e WonderWoman na lista de ativos e Batman na de inativos. Esta modificação podemos fazê-la como quisermos: ou programando as estruturas de dados e adicionar ao JSON ou fazer a alteração diretamente no JSON.
```python
# Completa o exercicio aqui
import json

super_herois = {
	"nomedaEquipa": "Super Hero Squad",
	"cidade": "Metro City",
	"formado": 2016,
	"baseSecreta": "Super Tower",
	"activo": "Sim",
	"membros": [
		{
			"nome": "SuperMan",
			"idade": 29,
			"identidadeSecreta": "Clart Kent",
			"poderes": [
				"Super força",
				"Super velocidade",
				"Raio laser"
			]
		},
		{
			"nome": "Batman",
			"idade": 350,
			"identidadeSecreta": "Bruce Wayne",
			"poderes": [
				"Detective",
				"Dinheiro",
				"Raio laser"
			]
		},
		{
			"nome": "Wonder Woman",
			"idade": 900,
			"identidadeSecreta": "Diana de Temiscira",
			"poderes": [
				"Super força",
				"Super velocidade"
			]
		}
	]
}

print(type(super_herois))
print(super_herois)
print(" ")

def buscar_herois(heroi = str):
    """
    função que retorna o herois pesquisado.
    heroi = nome do heroi a buscar
    OBS: criei essa função com o intuito de diminuir a quantidade de variaveis criadas para realização do exercicio 
    """
    for x in super_herois["membros"]:
        if super_herois["membros"][0]["nome"] == heroi:
            return super_herois["membros"][0]
            break
        elif super_herois["membros"][1]["nome"] == heroi:
            return super_herois["membros"][1]
            break
        elif super_herois["membros"][2]["nome"] == heroi:
            return super_herois["membros"][2]
            break
    
membrosAtivos = [buscar_herois("SuperMan"),buscar_herois("Wonder Woman")]

membrosInativos = [buscar_herois("Batman")]


super_herois["membros"].clear()


super_herois["membros"]=[{"membrosAtivos":membrosAtivos},{"membrosInativos":membrosInativos}]

#Convertendo em JSON
super_herois_json = json.dumps(super_herois)

print(" ")
print(type(super_herois_json))
print(super_herois_json)
```
