#Ejercicio de Indexacin y Seleccin de Datos en Pandas
#En este ejercicio practicars diferentes mtodos de indexacin y seleccin de datos en Pandas, tanto en objetos Series como DataFrames.

```python
import pandas as pd
import numpy as np

# PARTE 1: Trabajando con Series
# 1.1 Crea una Serie con los valores [10, 20, 30, 40, 50] y los ndices ['a', 'b', 'c', 'd', 'e']
serie = pd.Series([10, 20, 30, 40, 50],['a', 'b', 'c', 'd', 'e'])
print(serie)

# 1.2 Accede al valor correspondiente al ndice 'c' usando la notacin de diccionario
print(serie["c"])

# 1.3 Accede a los valores correspondientes a los ndices 'b' hasta 'd' utilizando slicing
print(serie["b":"d"])

# 1.4 Accede a los valores correspondientes a las posiciones 1 y 3 usando indexacin implcita (posiciones)
print(serie[1:4])

# 1.5 Crea una mscara booleana que seleccione los valores mayores a 25 y aplcala
print(serie[serie>25])

# 1.6 Utiliza indexacin fancy para seleccionar los valores correspondientes a los ndices 'a', 'c', 'e'
print(serie[["a","c","e"]])

# 1.7 Crea una nueva Serie con ndices numricos [1, 3, 5, 7, 9]
# y valores ['A', 'B', 'C', 'D', 'E']
new_serie = pd.Series(['A', 'B', 'C', 'D', 'E'], index=[1, 3, 5, 7, 9])
print(new_serie)

# 1.8 Extrae el valor en el ndice 5 usando .loc
print(new_serie.loc[5])

# 1.9 Extrae el valor en la posicin 2 (tercer elemento) usando .iloc
print(new_serie.iloc[2])
valor_iloc = new_serie.iloc[2]
valor_iloc

```

```python
# PARTE 2: Trabajando con DataFrames
# 2.1 Crea un DataFrame con los siguientes datos:
# - Columnas: 'nombre', 'edad', 'ciudad', 'salario'
# - Datos:
#   * Ana, 28, Madrid, 45000
#   * Pablo, 34, Barcelona, 55000
#   * Luca, 29, Valencia, 42000
#   * Mario, 42, Sevilla, 62000
#   * Sofa, 31, Madrid, 51000
datos = {'Nombre': ['Ana', 'Pablo', 'Luca', 'Mario', 'Sofa'],
    'Edad': [28, 34, 29, 42, 31],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid'],
    'Salario': [45000, 55000, 42000, 62000, 51000]}
df = pd.DataFrame(datos)
print(df)
# Establece la columna 'nombre' como ndice
df.set_index('Nombre', inplace=True)
print("\nDataFrame creado:")
print(df)
# 2.2 Accede a la columna 'ciudad' usando notacin de diccionario
print(df["Ciudad"])

# 2.3 Accede a la columna 'edad' usando notacin de atributo
print(f'Columna Edad:\n{df.Edad}')

# 2.4 Selecciona las filas desde la posicin 1 hasta la 3 (sin incluir la 4)
print(f'Filas posicion:\n{df.iloc[1:3]}')

# 2.5 Selecciona las filas correspondientes a 'Pablo' hasta 'Mario'
pablo_hasta_mario = df.loc["Pablo":"Mario"]
print(f'Filas Pablo hasta Mario:\n{pablo_hasta_mario}')

# 2.6 Aade una nueva columna 'experiencia' con los valores [3, 8, 4, 15, 6]
df["Experiencia"] = [3,8,4,15,6]
print(df)

# 2.7 Calcula y aade una columna 'salario_hora' que sea el salario dividido entre
# (experiencia * 500)
df["Salario_hora"]= df["Salario"] / (df["Experiencia"] * 500)
print("\nDataFrame con columna 'salario_hora':")
print(df)
# 2.8 Selecciona las filas donde el salario es mayor a 50000
print("\nDataFrame con salario > 50000:")
print(df[df['Salario'] > 50000])

# 2.9 Selecciona las columnas 'nombre' y 'salario' para las personas de Madrid
print("\nDataFrame con Ciudad Madrid:")
print(df.loc[df['Ciudad'] == 'Madrid', 'Salario'])

# 2.10 Utiliza .loc para seleccionar las filas de 'Pablo' a 'Mario' y las columnas 'edad' y 'ciudad'
print('2.10')
print(df.loc['Pablo':'Mario', ["Edad", "Ciudad"]])

# 2.11 Utiliza .iloc para seleccionar:
# - Primera y ltima fila
# - Segunda y tercera columna
seleccion_iloc = df.iloc[[0, -1], [1, 2]]
print("\nSeleccin con .iloc (primera y ltima fila, segunda y tercera columna):")
print(seleccion_iloc)

# 2.12 Modifica el salario de 'Luca' a 46000
df.loc['Luca', 'Salario'] = 46000
print(df)

# 2.13 Selecciona las filas donde el salario_hora es mayor que 12
print(df[df['Salario_hora'] > 12])
```

```python
# PARTE 3: Combinando tcnicas avanzadas
# 3.1 Ordena el DataFrame por salario de mayor a menor
print(df.sort_values("Salario",ascending=False))

# 3.2 Selecciona el nombre y la ciudad de las 2 personas con mayor salario
df_sort = df.sort_values("Salario",ascending=False)[:2]["Ciudad"]
print(df_sort)

# 3.3 Calcula y aade una columna 'categoria' que sea:
# - 'Junior' si experiencia < 5
# - 'Senior' si experiencia >= 5 y experiencia < 10
# - 'Experto' si experiencia >= 10
# Pista: utiliza np.select() o una funcin aplicada con .apply()
condiciones = [(df['Experiencia'] < 5),
    (df['Experiencia'] >= 5) & (df['Experiencia'] < 10),
    (df['Experiencia'] >= 10)]
categoria = [ "Junior","Senior","Experto"]
df["Categoria"] = np.select(condiciones,categoria,default="Desconhecido")
df
# 3.4 Agrupa el DataFrame por 'ciudad' y calcula el salario promedio por ciudad
salario_por_ciudad = df.groupby("Ciudad")["Salario"].mean()
salario_por_ciudad

# 3.5 Reemplaza los valores de 'Madrid' por 'MAD' y 'Barcelona' por 'BCN' en la columna 'ciudad'
df["Ciudad"] = df["Ciudad"].replace("Madrid","MAD")
df["Ciudad"] = df["Ciudad"].replace("Barcelona","BCN")
df
```

#Ejercicio 1: Deteccin de Valores Nulos


```python
import pandas as pd
import numpy as np

# 1.1 Crea una Serie con los siguientes valores: [10, None, 20, np.nan, 30, pd.NA]
serie = pd.Series([10, None, 20, np.nan, 30, pd.NA])
print(serie)

# 1.2 Utiliza el mtodo apropiado para identificar qu posiciones contienen valores nulos
print("Identificar valores nulos")
valores_nulos = serie.isnull()
print(valores_nulos)

# 1.3 Calcula qu porcentaje de los valores en la Serie son nulos
porcentagem_null = (valores_nulos.mean()*100 )
print("Porcentagem de nulos")
print(f'{porcentagem_null}%')

# 1.4 Crea un DataFrame con la siguiente informacin:
# - Columnas: 'A', 'B', 'C'
# - Datos:
#   * Fila 1: [1, np.nan, 3]
#   * Fila 2: [4, 5, np.nan]
#   * Fila 3: [np.nan, 8, 9]
#   * Fila 4: [10, None, 12]
df = pd.DataFrame([[1, np.nan, 3],[4, 5, np.nan],[np.nan, 8, 9],[10, None, 12]],columns=['A', 'B', 'C'])
print(df)

# 1.5 Identifica qu columnas contienen al menos un valor nulo
print(df.columns[df.isna().any()].tolist()) #.tolist() convierte el ndice de columnas filtrado (que es un objeto Index) en una lista.

# 1.6 Cuenta el nmero de valores nulos por columna
valores_nulos = df.isna().sum()
print(valores_nulos)

# 1.7 Cuenta el nmero de valores nulos por fila
valores_nulos_fila = df.isna().sum(axis=1)
print(valores_nulos_fila)
```

#Ejercicio 3: Rellenando Valores Nulos

```python
# 3.1 Crea una Serie de tiempo con fechas diarias para una semana y valores
# [10, np.nan, 15, np.nan, np.nan, 20, 25]
fechas = pd.date_range('2025-01-01', periods=7, freq='D')
series = pd.Series([10, np.nan, 15, np.nan, np.nan, 20, 25],index=fechas)
series

# 3.2 Rellena todos los valores nulos con cero
print("Rellena todos los valores nulos con cero")
series_notnull = series.fillna(0)
print(series_notnull)

# 3.3 Rellena los valores nulos con la media de los valores no nulos
series_media = series.fillna(series.mean())
print("Rellena los valores nulos con la media de los valores no nulos")
print(series_media)

# 3.4 Utiliza el mtodo forward fill para propagar el ltimo valor vlido
series_ffill = series.ffill()
print("Utiliza el mtodo forward fill para propagar el ltimo valor vlido")
print(series_ffill)

# 3.5 Utiliza el mtodo backward fill para propagar el siguiente valor vlido
series_bfill = series.bfill()
print("Utiliza el mtodo backward fill para propagar el siguiente valor vlido")
print(series_bfill)

# 3.6 Utiliza una combinacin de forward fill y backward fill para asegurarte
# de que no queden valores nulos (primero ffill, luego bfill)
series_combinada = series.ffill().bfill()
print(series_combinada)

# 3.7 Crea un DataFrame con datos de ventas mensuales para 3 productos
# con algunos valores faltantes
meses = pd.date_range("2025-01-01",periods=6,freq="ME")
ventas = pd.DataFrame({
    'Producto A': [100, 120, np.nan, 140, np.nan, 160],
    'Producto B': [200, np.nan, 220, np.nan, 240, 250],
    'Producto C': [np.nan, 300, 320, 340, 360, np.nan]
}, index=meses)
print(ventas)
# 3.8 Rellena los valores nulos de cada producto con la media de ventas de ese producto
valores_nulos = ventas.fillna(ventas.mean())
print(valores_nulos)

# 3.9 Rellena los valores nulos con el valor de la misma columna de la fila anterior
valores_rellenados_ffill = ventas.ffill()
valores_rellenados_ffill
```

```python
import pandas as pd
import numpy as np
rng = np.random.default_rng(42)
# 1.1 Crea un MultiIndex utilizando from_product con las siguientes categoras:
# - Regiones: 'Norte', 'Sur', 'Este', 'Oeste'
# - Aos: 2021, 2022
# - Trimestres: 'Q1', 'Q2', 'Q3', 'Q4'
regiones = ['Norte', 'Sur', 'Este', 'Oeste']
anos = [2021,2022]
trimestres = ['Q1', 'Q2', 'Q3', 'Q4']

multi_index = pd.MultiIndex.from_product([regiones,anos,trimestres],names=["regiones","aos","trimestres"])
print(multi_index)

# 1.2 Crea una Series con el MultiIndex anterior y datos aleatorios entre 100 y 1000 (enteros)
# para representar ventas trimestrales
# Pista: Necesitars 32 valores aleatorios (4 regiones  2 aos  4 trimestres)
serie = pd.Series(rng.integers(100,1000,32),index=multi_index)
print(serie.head())
# 1.3 Asigna nombres adecuados a los niveles del ndice: 'region', 'ao', 'trimestre'
serie.index.set_names(["Regiones","Aos","Trimestres"],inplace=True)
print(serie)

# 1.4 Accede a las ventas del 'Norte' para el ao 2022
print("ventas del 'Norte' para el ao 2022")
ventas_norte_2022 = serie.loc["Norte",2022]
print(ventas_norte_2022)
# 1.5 Accede a las ventas de todas las regiones para el primer trimestre (Q1) de 2021
# Alternativa usando IndexSlice
idx = pd.IndexSlice
print("ventas de todas las regiones para el primer trimestre (Q1) de 2021")
ventas_tri_2021 = serie.loc[idx[:, 2021, 'Q1']]
print(ventas_tri_2021)

# 1.6 Calcula el promedio de ventas por regin (independientemente del ao y trimestre)
print("Calcula el promedio de ventas por regin (independientemente del ao y trimestre)")
ventas_promedio = serie.groupby("Regiones").mean()

# 1.7 Reorganiza los datos usando unstack para convertir el nivel 'trimestre' en columnas
print("Reorganiza los datos usando unstack para convertir el nivel 'trimestre' en columnas")
serie_reorganizada = serie.unstack(level="Trimestres")
print(serie_reorganizada)

# 1.8 A partir del resultado anterior, convierte el nivel 'ao' en columnas
# para tener una vista ms compacta con la regin como nico ndice
serie_reorganizada2 = serie_reorganizada.unstack()
serie_reorganizada2
```

#Ejercicio 2: Manipulacin avanzada de MultiIndex en DataFrames

```python
import pandas as pd
import numpy as np
rng = np.random.default_rng(42)
# 2.1 Crea un DataFrame con los siguientes datos:
# - MultiIndex para filas con dos niveles: 'estudiante' y 'periodo'
#   - Estudiantes: 'Ana', 'Carlos', 'Elena'
#   - Perodos: 'Parcial1', 'Parcial2', 'Final'
# - MultiIndex para columnas con dos niveles: 'asignatura' y 'mtrica'
#   - Asignaturas: 'Matemticas', 'Ciencias', 'Historia'
#   - Mtricas: 'nota', 'asistencia'
# - Rellena el DataFrame con:
#   - Notas aleatorias entre 60 y 100
#   - Asistencia aleatoria entre 70% y 100%
estudiantes = ['Ana', 'Carlos', 'Elena']
periodos = ['Parcial1', 'Parcial2', 'Final']
multi_index_filas = pd.MultiIndex.from_product([estudiantes,periodos],names=["Estudiantes","Periodos"])
multi_index_filas

asignaturas= ['Matemticas', 'Ciencias', 'Historia']
metricas= ['nota', 'asistencia']
multi_index_columnas = pd.MultiIndex.from_product([asignaturas,metricas],names=['Asignatura' , 'Mtrica'])

notas = rng.integers(60,100,size=(len(multi_index_filas), len(multi_index_columnas) // 2))
asistencia = rng.integers(70,100,size=(len(multi_index_filas), len(multi_index_columnas) // 2))

datos = np.column_stack([notas, asistencia])
df = pd.DataFrame(datos,columns=multi_index_columnas,index=multi_index_filas)
print(df)
# 2.2 Accede a todas las notas de 'Matemticas' para todos los estudiantes y perodos
idx = pd.IndexSlice
print("NOTAS MATEMATICAS")
notas_matematica = df["Matemticas","nota"]
print(notas_matematica)
# 2.3 Obtn todas las mtricas de 'Ana' para el 'Parcial1'
metrica_ana = df.loc[('Ana', 'Parcial1'), :]
print(f'Las metricas de Ana para el Parcial1 son: {metrica_ana}')

# 2.4 Calcula el promedio de notas por estudiante en todas las asignaturas
# (independientemente del perodo)
idx = pd.IndexSlice
notas_df = df.loc[:,idx[:,"nota"]]
promedio_notas = notas_df.groupby("Estudiantes").mean()
promedio_notas

# 2.5 Calcula el promedio de asistencia por asignatura
# (independientemente del estudiante y perodo)
asistencia_df = df.loc[:, idx[:, 'asistencia']]
promedio_asistencia_asignatura = asistencia_df.mean(axis=0)
print("\nPromedio de asistencia por asignatura:")
print(promedio_asistencia_asignatura)

# 2.6 Encuentra al estudiante con la mejor nota final en 'Ciencias'
mejor_estudiante_ciencias = df.loc[idx[:,"Final"],idx["Ciencias","nota"]]
print(mejor_estudiante_ciencias)
mejor_nota = mejor_estudiante_ciencias.max()
print(f'El mejor estudiante fue {mejor_estudiante_ciencias.idxmax()[0]} con nota {mejor_nota}')


# 2.7 Reorganiza el DataFrame para tener un formato donde:
# - Las filas sean los estudiantes
# - Las columnas sean una combinacin de asignatura, perodo y mtrica
# Pista: usa un mtodo stack/unstack adecuado
nuevo_df = df.unstack(level=0)
print(nuevo_df)

# 2.8 A partir de tu DataFrame original, reinicia el ndice para obtener un DataFrame
# "plano" donde 'estudiante' y 'periodo' sean columnas normales
plano = df.reset_index()
print(plano)

# 2.9 A partir del DataFrame "plano", vuelve a establecer el ndice multinivel original
# usando set_index
plano_multinivel = plano.set_index(['Estudiantes', 'Periodos'])
plano_multinivel
```

#Ejercicios Concat

```python
import pandas as pd
import numpy as np

# 1.1 Crea dos Series:
# - ser1 con valores ['A', 'B', 'C'] e ndices [1, 2, 3]
# - ser2 con valores ['D', 'E', 'F'] e ndices [4, 5, 6]
ser1 = pd.Series(['A', 'B', 'C'],index=[1,2,3])
ser2 = pd.Series(['D', 'E', 'F'],index=[4,5,6])

# 1.2 Concatena las dos Series verticalmente (por defecto)
ser_concat = pd.concat([ser1,ser2])
ser_concat

# 1.3 Crea dos DataFrames:
# - df1 con columnas 'nombre' y 'edad', y dos filas con datos:
#   ('Ana', 25) y ('Carlos', 30)
# - df2 con las mismas columnas y dos filas:
#   ('Elena', 35) y ('David', 28)
column_df =["Nombre","Edad"]
row_df1 = [('Ana', 25), ('Carlos', 30)]
row_df2 = [('Elena', 35) , ('David', 28)]
df1 = pd.DataFrame(row_df1,columns=column_df)
print(df1)
df2 = pd.DataFrame(row_df2,columns=column_df)
print(df2)

# 1.4 Concatena los dos DataFrames verticalmente
df_concat = pd.concat([df1,df2])
print(df_concat)

# 1.5 Crea dos DataFrames ms:
# - df3 con columnas 'nombre' y 'ciudad', y datos:
#   ('Ana', 'Madrid') y ('Carlos', 'Barcelona')
# - df4 con columnas 'nombre' y 'profesin', y datos:
#   ('Ana', 'Ingeniera') y ('Carlos', 'Mdico')
df3 = pd.DataFrame({"Nombre":["Ana","Carlos"],
                    "Ciudad": ["Madrid","Barcelona"]})
df4 = pd.DataFrame({"Nombre":["Ana","Carlos"], "Profesin": ["Ingeniera","Mdico"]})
print(df3)
print(df4)
# 1.6 Concatena df3 y df4 horizontalmente (por columnas)
df_concat_df3_df4 = pd.concat([df3,df4],ignore_index=False,axis="columns")
df_concat_df3_df4

# 1.7 Verifica qu ocurre si concatenas DataFrames con ndices duplicados
# Crea df5 idntico a df1 pero estableciendo sus ndices a [0, 1]
# Crea df6 idntico a df2 pero estableciendo sus ndices a [0, 1]
# Concatena df5 y df6 verticalmente
df5 = pd.DataFrame({
    'nombre': ['Ana', 'Carlos'],
    'edad': [25, 30]
}, index=[0, 1])
df6 = pd.DataFrame({
    'nombre': ['Elena', 'David'],
    'edad': [35, 28]
}, index=[0, 1])
try:
  df7 = pd.concat([df5,df6],verify_integrity=True)
  print(df7)
except:
  df7 = pd.concat([df5,df6],ignore_index=True)
  print(df7)

# 1.8 Utiliza el parmetro verify_integrity=True al concatenar df5 y df6
# y maneja el error que se produce
try:
  df7 = pd.concat([df5,df6],verify_integrity=True)
  print(df7)
except:
  df7 = pd.concat([df5,df6],ignore_index=True)
  print(df7)

# 1.9 Concatena df5 y df6 ignorando los ndices (usando ignore_index=True)
df7 = pd.concat([df5,df6],ignore_index=True)
print(df7)

# 1.10 Concatena df5 y df6 aadiendo una clave para identificar el origen de cada fila
# (usando el parmetro keys=['primero', 'segundo'])
df8_concat = pd.concat([df5,df6],keys=["Primeiro","Segundo"])
print(df8_concat)
```

```python
# 2.1 Crea los siguientes DataFrames:
# - df_ventas_q1: columnas 'producto', 'unidades', 'ingresos' con 3 productos distintos
# - df_ventas_q2: mismas columnas que df_ventas_q1 con 3 productos (algunos repetidos de Q1)
# - df_costos_q1: columnas 'producto', 'costos', 'proveedor' con los mismos productos que df_ventas_q1
df_ventas_q1 = pd.DataFrame({'Producto': ["ternera","Cerdo","Pollo"],
                             'Unidades': [4,5,6],
                             'Ingresos': [10,12,14]})
df_ventas_q2 = pd.DataFrame({'Producto': ["Tomate","Cerdo","Pollo"],
                             'Unidades': [2,5,6],
                             'Ingresos': [5,12,14]})
df_costos_q1 = pd.DataFrame({'Producto': ["ternera","Cerdo","Pollo"],
                             'Costos': [2,5,6],
                             'Proveedor': ["Mercadona","Spar","Lidel"]})

print(df_ventas_q1)
print(df_ventas_q2)
print(df_costos_q1)


# 2.2 Concatena df_ventas_q1 y df_ventas_q2 verticalmente
# para obtener los datos de ventas de ambos trimestres
ventas_q1_q2 = pd.concat([df_ventas_q1,df_ventas_q2])
print(ventas_q1_q2)

# 2.3 Concatena df_ventas_q1 y df_costos_q1 horizontalmente
# Qu ocurre con las columnas? Se conservan todas?
ventas_costos_q1 = pd.concat([df_ventas_q1,df_costos_q1],axis=1)
print(ventas_costos_q1)#Se conservam y inclui valor NaN en las lineas vacias

# 2.4 Ahora realiza la misma concatenacin pero solo manteniendo
# las columnas comunes (usando join='inner')
ventas_costos_q1 = pd.concat([df_ventas_q1,df_costos_q1],join="inner")
print(ventas_costos_q1)

# 2.5 Crea un nuevo DataFrame df_q3 con datos de un tercer trimestre
# pero con una columna adicional 'descuento' y sin la columna 'ingresos'
df_q3 = pd.DataFrame({'Producto': ["ternera","Cerdo","Pollo"],
                             'Unidades': [4,5,6],
                             'Descuento': [2,3,5]})
print(df_q3)
# 2.6 Concatena df_ventas_q1, df_ventas_q2 y df_q3 verticalmente.
# Qu valores se generan para las columnas que faltan?
ventas_concat = pd.concat([df_ventas_q1,df_ventas_q2,df_q3])
ventas_concat #generam valores NaN para los datos faltantes en ingreso y descuento

# 2.7 Concatena df_ventas_q1 y df_costos_q1, pero antes reindexando
# df_costos_q1 para que tenga las mismas columnas que df_ventas_q1
# (usa el mtodo reindex)
costos_reindexado = ventas_costos_q1.reindex(columns=df_ventas_q1.columns)
ventas_costos_reindexado = pd.concat([df_ventas_q1, costos_reindexado])
print("\nVentas y costos con reindexacin:")
print(ventas_costos_reindexado)


# 2.8 Concatena verticalmente df_ventas_q1 y df_ventas_q2,
# pero aade un nivel de ndice para identificar el trimestre
ventas_concat = pd.concat([df_ventas_q1,df_ventas_q2],keys=["Q1","Q2"])
print(ventas_concat)

# 2.9 Concatena horizontalmente df_ventas_q1 y df_costos_q1,
# pero aadiendo un nivel de columna para identificar si los datos
# son de 'ventas' o 'costos'
ventas_concat = pd.concat([df_ventas_q1,df_ventas_q2],keys=["Ventas","Costos"],axis=1)
print(ventas_concat)
```
